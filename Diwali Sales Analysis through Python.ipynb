```python
boxplot = sns.countplot(x='Gender', data=df)
boxplot.set_xlabel('Gender')
boxplot.set_ylabel('Frequency')
for bars in boxplot.containers:
    boxplot.bar_label(bars)
#to find that how many male and female are there in total
```

```python
barplot = sns.barplot(data=df, x="Marital_Status", y="Zone")
sns.set(rc={'figure.figsize':(10, 6)}) 
for bars in barplot.containers:
    barplot.bar_label(bars)
#to find that the how many people are married as per Zone
```

```python
lineplot = sns.lineplot(data=df, x="Occupation", y="Amount")
sns.set(rc={'figure.figsize':(20, 10)}) 
plt.xticks(fontsize=10)
# to find the total amount according to the occupation
```

```python
df['Product_Category'].unique()
```

```python
hist = sns.histplot(x="Product_Category", data=df, kde=True, hue="Product_Category")
sns.set(rc={'figure.figsize':(40, 10)}) 
for bars in hist.containers:
    hist.bar_label(bars)
plt.xticks(fontsize=15)
#to find how many product category is there in each zone
```

```python
statebar = sns.barplot(data=df, x="State", y="Amount")
for bars in statebar.containers:
    statebar.bar_label(bars)
plt.xticks(fontsize=15)
# to find the sum of amount as per state
```

```python
#to find the no. of orders per state
box_plot = sns.boxplot(x="State", y="Orders", data=df)
```

```python
boxplot = sns.countplot(x='Marital_Status', data=df, hue="Age Group")
boxplot.set_xlabel('Martial Status')
boxplot.set_ylabel('Frequency')
for bars in boxplot.containers:
    boxplot.bar_label(bars)
plt.xticks(fontsize=15)
#to find the how many people are married according to age group
```
